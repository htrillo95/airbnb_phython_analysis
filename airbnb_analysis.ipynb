## Objective 1: Data Prep & QA

- Read in the AirBnB listings data (use `low_memory=False` and `encoding="ISO-8859-1"` in `read_csv`)
- Cast any date columns as a datetime format
- Filter the data down to just listings in the city of Paris
- QA the Paris listings data: check for missing values, and calculate the minimum, maximum, and average for each numeric field

```python
import pandas as pd

listings = pd.read_csv( ##Read in the AirBnB listings data 
    "Listings.csv",
    encoding= "ISO-8859-1",
    low_memory=False,
    parse_dates=["host_since"] ##Cast any date columns as a datetime format
)
```

```python
listings.head()
```

```python
listings.info()
```

```python
paris_listings = (
    listings
        .query("city == 'Paris'") ## Filter the data down to just listings in the city of Paris
        .loc[:, ["host_since", "neighbourhood", "city", "accommodates", "price"]] ## Filters down further to columns we want
)

paris_listings.info()
```

```python
paris_listings.isna().sum() ## Returns true is a value is missing (ie "host_since" here) 
                            ## add .sum to get sum of values missing
```

```python
paris_listings.describe() ##check for missing values, calculate min, max, avg for each numeric field
                            ## gives statistics of interest 
                          ## minimum value for 'accommodates' & 'price' = 0 (does not make sense)
```

## Objective 2: Prepare for Visualization

- Create a DataFrame called `paris_listings_neighbourhood`.  
  Group the Paris listings by `neighbourhood` and calculate the average price for each.  
  Sort by price in ascending order.

- Create a DataFrame called `paris_listings_accommodations`.  
  Filter your data down to the most expensive neighbourhood in Paris.  
  Group it by `accommodates` and calculate the average price for each.  
  Sort by price in ascending order.

- Finally, create a DataFrame called `paris_listings_over_time`.  
  Group the data by the year component of `host_since`.  
  Calculate a count of rows to get the number of new hosts for each year,  
  and the average price of listings for each year.

```python
paris_listings_neighbourhood = ( ##Create DataFrame
    paris_listings
    .groupby("neighbourhood") ##Group by `neighbourhood` 
    .agg({"price": "mean"})  ##calculate avg price for each 
    .sort_values("price")    ##Sort by price in ascending order
)

paris_listings_neighbourhood.tail()
```

```python
paris_listings_accommodates = (  ##Create DataFrame
    paris_listings
    .query("neighbourhood == 'Elysee'") ##Filter down to most expensive neighbourhood (used prior script result - Elysee = 210 - most expensive)
    .groupby("accommodates")  ##Group by `accommodates`
    .agg({"price": "mean"})   ##Calculate the average price for each
    .sort_values("price")    ##Sort in ascending order
    
)

paris_listings_accommodates.head()
```

```python
paris_listings_over_time = (   ##Create DataFrame
    paris_listings
    .set_index("host_since")  ##Group by year component `host_since`
    .resample("YE")
    .agg({
        "neighbourhood": "count",   ##Calculate count of rows = new hosts for each year
        "price": "mean"             ##Average price of listings for each year
    })
)

paris_listings_over_time.head()
```

## Objective 3: Visualize the Data

- Build a horizontal bar chart of average rent price by neighbourhood. Which neighbourhoods stand out?
- Build a horizontal bar chart of average price by accommodates in the most expensive neighbourhood. Are the results intuitive?
- Finally, build line charts of new hosts per year and average price by year. What happened to new hosts in 2015? Was average price impacted?
- **Challenge:** Plot both time series in a dual-axis line chart!

```python
import seaborn as sns

(paris_listings_neighbourhood
    .plot
    .barh(  ##horizontal bar chart of avg rent price by neighbourhood
        title="Average Listing Price by Paris Neighbourhood",
        xlabel="Price Per Night (Euros)",
        ylabel="Neighbourhood",
        legend=None
    )
)

sns.despine()
```

```python
(paris_listings_accommodates
    .plot
    .barh(   ##horizontal bar chart of avg price by accommodates in most expensive neighbourhood
        title="Average Listing Price by Accommodation Number",
        xlabel="Price Per Night (Euros)",
        ylabel="Accommodation Capacity",
        legend=None
    )
)

sns.despine()
```

```python
paris_listings_over_time["neighbourhood"].plot( ##line charts of new hosts per year (time-series)
    ylabel="New Hosts",
    title="New AirBnB Hosts in Paris Over Time"

)

    
```

```python
paris_listings_over_time["price"].plot(  #line charts of avg price by year (time-series)
    ylabel="Average Price (Euros)",
    title="Average AirBnB Price in Paris Over Time"

)
```

```python
import matplotlib.pyplot as plt
                        ##Plot both time series in a dual-axis line chart
fig, ax = plt.subplots()

ax.plot(
    paris_listings_over_time.index,
    paris_listings_over_time["neighbourhood"],
    label="New Hosts",
    c="pink"
)

ax.set_ylabel("New Hosts")

ax2 = ax.twinx()

ax2.plot(
    paris_listings_over_time.index,
    paris_listings_over_time["price"],
    label="Average Price",

)
ax2.set_ylim(0)

ax2.set_ylabel("Average Price")

ax.set_title("2015 Reulations Lead to Fewer New Hosts, Higher Prices")
```
